```python
from glob import glob
imgList = glob('images/*')  # 이미지에 있는 파일목록 불러오기 (경로설정)
labelList = glob('annotations/*')   # 어노테이션에 있는 파일목록 불러오기 (경로설정)
```

```python
len(imgList), len(labelList)
```

```python
imgnameList = []    #확장자를 제외한 파일 이름만 남김
for jpgname in imgList:
    imgname = jpgname.split('\\')[1].split('.')[0]
    imgnameList.append(imgname)
imgnameList
```

```python
labelnameList = []  # 확장자를 제외한 파일 이름만 남김
for name in labelList:
    labelname = name.split('\\')[1].split('.')[0]
    labelnameList.append(labelname)
labelnameList
```

```python
len(imgnameList), len(labelnameList)
```

```python
for name in imgnameList:
    if name in labelnameList:   # 두 파일의 이름이 같으면 pass
        pass
    else:       # 두 파일 이름이 다르면 이름 출력
        print(name)
```
